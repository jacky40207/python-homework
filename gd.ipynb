```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sympy import *
import math
import random
from ipywidgets import interact
```

```python
x=np.linspace(0,50,200)

def f(x): 
    return (x-6)*np.cos(x)+(x-40)*np.sin(x)
y=f(x)
plt.plot(x, y)
```

```python
start = random.uniform(0,50)
lr = 0.05
plt.plot(x,y)
for i in range(50):
    start = start - lr * (start - 40)*np.cos(start)
    plt.scatter(start,f(start))
```
