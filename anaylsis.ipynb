```python
# basic
import numpy as np
import pandas as pd

# get data
import pandas_datareader as pdr

# visual
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

#time
import datetime as datetime
```

```python
start = datetime.datetime(2018,4,19)
campany = [ '3045.TW','2412.TW']
df_stock = pdr.DataReader(campany, 'yahoo', start=start)
```

```python
adjClose = df_stock['Adj Close']
adjClose.plot()
```

```python
df_stock.head()
```

```python
plt.rcParams['axes.unicode_minus']=False
adjClose_pct = adjClose.pct_change()
sns.jointplot('2412.TW','3045.TW',adjClose_pct, kind="scatter")
```

```python
sns.regplot('2412.TW','3045.TW',adjClose_pct)
```

```python
c=df_stock["Adj Close"]["2412.TW"]
t=df_stock["Adj Close"]["3045.TW"]
```

```python
rc=c.diff()/c
rt=t.diff()/t
```

```python
rc.plot()
rt.plot()
```

```python
c.rolling(window=5).mean().plot()
t.rolling(window=5).mean().plot()
```
